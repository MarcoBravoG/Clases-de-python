### ESTRUCTURA DE REPETICIÓN FOR - PARA

La sentencia for de Python itera sobre los ítems de cualquier secuencia, en el orden que aparecen en la misma.

**Iterar sobre la función range() - rango**

En lugar de ser una función, range() en realidad es un tipo de secuencia inmutable. La secuencia resultante tendrá una lista desde el límite inferior, p. ej. cero, hasta el límite superior menos uno.

Por defecto el límite inferior o índice inicial será cero.

#### Sintaxis Diagramas de Flujo:


#### Sintaxis Python:
***<span style="color:blue"> for </span>*** + ***<span style="color:orange"> nombre_variable </span>*** + ***<span style="color:blue"> in range </span>*** + ***<span style="color:orange"> (parámetro1,parámetro2, parámetro3) </span>***

***Donde:***

<span style="color:green"> parámetro_1 </span> corresponde al valor inicial que tendrá la variable contador. El valor por defecto que se asigna es cero.

<span style="color:green">parámetro_2</span> corresponde al número de repeticiones ***menos 1*** que se ejecutará la estructura repetitiva.

<span style="color:green">parámetro_3</span> corresponde al incremento del contador, por defecto el incremento es 1.

***La estructura repetitiva al menos requiere que se coloque un parámetro, mismo que hace referencia el numero de iteraciones o repeticiones que realizará el bucle***

Por ejemplo:

for i in range(10):
    print(i)


```python
for i in range(5):
    print(i)
```

```python
for i in range(1,6):
    print(i)
```

```python
for i in range(1,8,3):
    print(i)
```

### Ejemplo 4:
Diseñe un algoritmo que solicite al usuario ingresar dos valores enteros: un límite inferior y un límite superior. El programa debe imprimir todos los números comprendidos entre estos límites. Asegúrese de incluir las validaciones necesarias para garantizar que el límite inferior sea menor que el límite superior.

```python
li = int(input("Ingrese el límite inferior: "))
ls = int(input("Ingrese el límite superior: "))

while li >= ls:
    print("El límite inferior debe ser menor que el superior.")
    li = int(input("Ingrese el límite inferior: "))
    ls = int(input("Ingrese el límite superior: "))

for dive in range(li, ls + 1):
    print(dive)
```

### Ejemplo 5:

Desarrolle un programa que determine el número mayor entre una cantidad de 𝑛 valores.

El programa debe solicitar al usuario ingresar la cantidad de números (𝑛) y, a continuación, los valores correspondientes para identificar el mayor de ellos.

```python
n = int(input("Ingrese la cantidad de números: "))

mayor = None

for dive in range(n):
    num = int(input(f"Ingrese el número {dive+1}: "))
    if mayor is None or num > mayor:
        mayor = num

print("El número mayor es:", mayor)
```

### Ejemplo 6:
Se requiere programar un generador de tablas de multiplicar que calcule los productos desde el número 1 hasta el número 8. Sin embargo, el usuario deberá especificar hasta qué número desea extender las multiplicaciones en cada tabla.
- 1 * 1 = 1
- 1 * 2 = 2
- 1 * 3 = 3
- 1 * 4 = 4
- 1 * 5 = 5
- 1 * 6 = 6
- 1 * 7 = 7
- 1 * 8 = 8

```python
limite = int(input("Numero a multiplicar: "))

for dive in range(1, 9):
    print(f"\nTabla del {dive}")
    for i in range(1, limite + 1):
        print(f"{dive} * {i} = {dive * i}")
```
