### <span style="color:orange"> DEFINICIÓN </span>

#### ¿QUÉ ES UNA ESTRUCTURA DE CONTROL REPETITIVA?
Una estructura de control repetitiva es aquella que permite ejecutar una serie de procesos de manera controlada n cantidad de veces.

Existen estructuras repetitivas que son:

1. ### <span style="color:orange"> ***Controlodas por un contador (definidos)*** </span><div style="text-align: justify">
    En estas estructuras de control se conoce el número de iteraciones que se va a ejecutar.
    Dentro de estas estructuras de control se encuentra la estructura de control ***Para (for)***, aquí claramente se conoce el número de iteraciones que se va a ejecutar </div>
    
    
2. ### <span style="color:orange">***Controladas  por una condición (indefinidos)***</span><div style="text-align: justify">
    Estas estructuras de control se ejecutarán siempre y cuando la condición de entrada que permita acceder a la estructura sea verdadera.</div>
    Dentro de estas estructuras de control se encuentran las estructura de control:
    1. <span style="color:green">***Mientras***</span> en Python **while**
    
    2. <span style="color:green">***Repetir Hasta***</span> esta estructura no maneja python
    
<div style="text-align: justify">Es importante considerar que, si no se garantiza que la condición que permitió ejecutar la estructura repetitiva en algún momento se vuelva falsa, la estructura no finalizará su ejecución y se convertirá en un bucle indefinido.</div>

### Ejercicio 5:

### Ejercicio 6:
Crear un algoritmo que calcule la suma de los primeros ***n*** números pares.

***Por ejemplo:*** Si ***n= 5***, la salida a presentar sería: <span style="color:green">***2 + 4 + 6 + 8 + 10 = 30***</span>

```python
I = int(input("Ingrese la cantidad de numeros pares: "))
suma = 0
contador = 1

while contador <= I:
    terminopar = 2 * contador
    suma = suma + terminopar
    print(terminopar, end=" ")
    if contador < I:
        print("+", end=" ")
    contador = contador + 1

print("= ",suma)
```

### Ejercicio 8:
Crear un algoritmo que calcule el factorial de los ***n*** primeros números.

***Por ejemplo:*** Si ***n= 5*** la salida a presentar sería: <span style="color:green">***5 x 4 x 3 x 2 x 1 = 120***</span>

```python
 n = int(input("Ingrese el valor de n: "))
factorial = 1
i = n  

print("Serie: ", end="")
while i >= 1:
    print(i, end="")
    if i > 1:
        print(" x ", end="")
    factorial *= i
    i -= 1  

print(f" = {factorial}")
```

### Ejercicio 9:
Crear un algoritmo que calcule la suma de los primeros ***n*** términos de la serie 1 - 2 + 3 - 4 + 5.

***Por ejemplo:*** Si ***n= 5*** la salida a presentar sería: <span style="color:green">***1 - 2 + 3 - 4 + 5 = 3***</span>

```python
n = int(input("Ingrese el valor de n: "))
suma = 0
i = 1  


print("Serie: ", end="")
while i <= n:
    print(i, end="")
    if i < n:
        
        print(" - " if (i + 1) % 2 == 0 else " + ", end="")
    
    suma += i if i % 2 != 0 else -i
    i += 1  


print(f" = {suma}")
```

### Ejercicio 10:
Crear un algoritmo que determine si un número ingresado por el usuario es un número primo.

```python
numero = int(input("Ingrese un número: "))
es_primo = True
i = 2

if numero <= 1:
    es_primo = False
else:
    while i * i <= numero:
        if numero % i == 0:
            es_primo = False
            break
        i = i + 1

if es_primo:
    print(f"{numero} es primo")
else:
    print(f"{numero} no es primo")
```
